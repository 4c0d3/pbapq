```python
def desastre():
    
    t = int(input('Qual a temperatura do reator (ºC)? '))
    while t > 30:
        if t <= 150:
            print('ok, tudo tranquilo!')
            t = int(input('Qual a temperatura do reator (ºC)? '))
        else:
            while t > 150:
                print('Alarme aceso! Cuidado!!')
                if t <= 180:
                    t = int(input('Qual a temperatura do reator (ºC)? '))
                while t > 180:
                    print('Alivia a pressão!!')
                    if t <= 190:
                        t = int(input('Qual a temperatura do reator (ºC)? '))
                    while t > 190:
                        print('Aumenta a vazão do trocador')
                        if t <= 200:
                            t = int(input('Qual a temperatura do reator (ºC)? '))
                        while t > 200:
                            r = input('O reator tem costura no topo?(s/n) ')
                            if r == 's':
                                print ('menos mal')
                                return 'reator derreteu! Sem reator não tem o que controlar, dia livre!!!!'
                            else:
                                print ('corre que o reator vai decolar!!!!')
                                return 'reator voador, salve-se quem puder!!!!'
    if t <= 30:
        return 'fim da batelada, vou pra casa!'
```

```python
temperaturas = [300 + 3*x for x in range(1, 100)] # K
pressoes = [1e5*x for x in range(1, 100)] # Pa
n = 1 # mol
R = 8.314 # J/mol/K 
```

```python
def calculaVolumes(temperaturas, pressoes):
    
    volumes = []
    indexT = 0
    while indexT < len(temperaturas):
        linha = []
        
        for pressao in pressoes:
            temperatura = temperaturas[indexT]
            volume = n*R*temperatura/pressao            
            linha.append(volume)
        
        volumes.append(linha)
        indexT += 1
    
    return volumes
```

```python
def calculaVolumes(temperaturas, pressoes):
    
    volumes = []
    index = 0
    
    while index < len(temperaturas):
        linha = []
        
        while index2 < len(pressoes):
            T = temperaturas[index]
            P = pressoes[index2]
            volume = n*R*T/P
            linha.append(volume)
            
            index2 += 1
        
        volumes.append(linha)
        
        index += 1
    
    return volumes
```

```python
def calculaVolumes(temperaturas, pressoes):
    
    volumes = []
    index = 0
    
    while index < len(temperaturas):
        linha = []
        
        for P in pressoes:
            T = temperaturas[index]
            volume = n*R*T/P
            linha.append(volume)
        
        volumes.append(linha)
        
        index += 1
    
    return volumes
```

```python
volumes = calculaVolumes(temperaturas, pressoes)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from mpl_toolkits.mplot3d.axes3d import Axes3D
import numpy as np

temperaturas_grid, pressoes_grid = np.meshgrid(temperaturas, pressoes)
volumes_grid = np.array(calculaVolumes(temperaturas, pressoes)).T

fig = plt.figure(figsize=(20,6))    
ax1 = fig.add_subplot(1, 2, 1, projection='3d')
p1 = ax1.plot_surface(temperaturas_grid, pressoes_grid, volumes_grid, cmap=mpl.cm.plasma, rstride=1, cstride=1, linewidth=0)
ax1.set_xlabel('T(K)')
ax1.set_ylabel('P(Pa)')
ax1.set_zlabel('V(m³)')
ax1.view_init(15, 20)
```
