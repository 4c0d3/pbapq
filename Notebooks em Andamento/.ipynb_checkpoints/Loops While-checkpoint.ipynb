## Problema 1:

O cálculo da porosidade de um meio é um parâmetro importante e assuntos relacionados as etapas do processo que envolvem transformação física da matéria dentro dele (operações unitárias, como filtração para recolher particulado sólido de um fluido). A porosidade ($\epsilon$) é um parâmetro adimensional que nos informa sobre a proporção preenchida de sólido e a proporção preenchida de fluido (como ar ou água), para um dado volume de controle. No dado contexto, o volume de controle será um meio filtrante ou uma torta de filtração, por exemplo.

$$ \epsilon = { V_{fluido} \over V_{sólido} + V_{fluido} } $$


No processo de fluidização, existe uma porosidade de mínima fluidização $\epsilon_{mf}$, relacionada ao desenho de um processo que trabalhe numa faixa de velocidade de escoamento tal que a fluidização é capaz de ocorrer. Ela pode ser obtida por uma correlação empírica quando a esfericidade ($\phi$) dos sólidos do leito fluidizado é conhecido, pela seguinte relação:

$$ {1 - \epsilon_{mf} \over \phi^2 \cdot \epsilon_{mf}^3} \approx 11 $$

Pela primeira equação, $\epsilon$ é obviamente um valor entre 0 e 1. Desenvolva um algoritmo para o método [Newton-Raphson](https://pt.wikipedia.org/wiki/M%C3%A9todo_de_Newton%E2%80%93Raphson) para encontrar os valores de $\epsilon$, quando $\phi = 0.8$. *Teste também para os valores de esfericidades encontrados no Problema 5 do notebook de funções (semana 2).

O método Newton-Raphson tem o objetivo de encontrar as raízes de uma função de maneira iterativa. A raíz de uma função é um novo valor de $x$ tal que $f(x) = 0$. Novos valores de $x$ são obtidos pelas informações do valor anterior, como segue:

$$x_{n+1} = x_{n} - {f(x_n) \over f'(x_n)}$$

$x$ irá se aproximar da raíz da função gradualmente, conforme sugere a figura abaixo. Para entender a lógica do método, clique no link acima: é bem simples.



- Notas e sugestões:

*Estabeleça uma tolerância para o seu método, para quando $f(x) \lessapprox tolerância$ ele assumir que esse $x$ é a raíz. Do contrário, ele vai entrar em loop infinito, porque ele vai chegar muito perto de 0, mas o computador não vai reconhecer o número infinitesimal como tipo "int" de valor 0.*

*Se você achar conveniente obter o módulo de $f(x)$, você pode obtê-lo com $abs(f(x))$.*

*Cria uma função que calcula a derivada, que será chamada pelo sua função que executa o método. Para implementar sua função derivada, use a definição da derivada como um limite.*

$$ {df \over dt} = \lim_{h \to 0} {f(t+h) - f(t) \over h} $$

```python
def derivadaPontual(f, t, h):
    dfdt = (f(t+h)-f(t))/h
    return dfdt

def newtonRaphson(f, t, h, tolerancia):
    
    while abs(f(t)) >= tolerancia:
        
        print(f(t))
        
        dfdt = derivadaPontual(f, t, h)
        t = t - f(t)/dfdt
            
    return t    
```

```python
esfericidades = [0.6713057485808569, 0.8059959770082346, 0.8455775086617149, 0.9104485400925976, 0.9393221042002651]

testa_porosidade = (lambda porosidade_mf, esfericidade = esfericidades[0]: 
                    (1 - porosidade_mf)/(esfericidade**2 * porosidade_mf**3) - 11)

newtonRaphson(testa_porosidade, 0.5, 0.0001, 0.0001)
```

## Problema 2:

Ao tentarmos usar apenas o eletromagnetismo, desconsiderando a mecânica quântica, para encontrar uma função que relaciona a energia potencial de dois átomos de hidrogênio que se aproximam, uma função cujo gráfico é surpreendentemente parecido com os valores experimentais é obtido.

Nesse modelo, os átomos vem do infinito, de encontro um ao outro. Conforme eles vão se aproximando, os átomos vão girando (pense aqui no elétron como **apenas** uma partícula que **orbita** o núcleo), de forma a minimizar as repulsões núcleo-núcleo e elétron-elétron ao passo que tenta maximizar a atração núcleo-elétron.

A função abaixo descreve esse comportamento.

$$ E = ckq^2 \Bigg [{ { (a + x) \cdot \sqrt { a^2 + x^2 } } - 8xa \over x \sqrt { a^2 + x^2 } } - 1 \Bigg ] $$

$E$ é a energia potencial do sistema.  
$c$ é a velocidade da luz no vácuo.  
$k$ é a permissividade elétrica no vácuo.  
$q$ é a carga de um elétron.  
$a$ é o valor de uma relação geométrica obtida na dedução. Assuma $a = 1 Å = 10^{-10}m $  
$x$ é a distância entre os dois núcleos dos átomos.

***Nota: essa função não descreve a realidade. O interesse nela é apenas a aparência da curva com a curva real, obtida pela mecânica quântica.***

Abaixo segue o plot do gráfico.

```python
f = lambda x, a = 1e-10, c = 3e8, q = 1.6e-19, k = 9e9: c*k*q**2*(((a+x) * (a**2+x**2)**0.5 - 8*x*a)/ x / (a**2+x**2)**0.5 - 1)
eixo_x = np.linspace(1e-11, 10e-10, 100)
eixo_y = list(map(f, eixo_x))
grafico = plt.plot(eixo_x, eixo_y)
```

Crie um algoritmo Hooke-Jeeves unidimensional para achar o fosso de energia da curva (o ponto de mínimo), bem como a distância(x) em que ele ocorre. Seria em torno deste ponto que o comprimento de ligação dos dois átomos oscilaria, ao formarem uma ligação química, porque a energia potencial nesse estado é a menor.

```python
import math as m

def hooke_jeeves_1d(f, chute, incremento, tolerancia, lim_inf, lim_sup):
            
    while incremento > tolerancia:
        
        f1 = f(chute)
        
        if chute + incremento <= lim_sup:
            f2 = f(chute + incremento)
        else:
            #f2 = 99999999999
            f2 = m.inf #infinito

        if chute - incremento >= lim_inf:
            f3 = f(chute - incremento)
        else:
            #f3 = 99999999999
            f3 = m.inf
        #colocar a função como infinito é apenas um artifício simples p/ que ela não seja escolhida, caso passe da fronteira.
        
    #se vc quiser transformar a função p/ achar um máximo, em vez de um mínimo, apenas troque os sinais relacionais aqui.
    #e troque "m.inf" por "-m.inf".
        if f1 < f2 and f1 < f3:
            incremento /= 2
        else:
            if f2 < f3:
                chute += incremento
            else:
                chute -= incremento
    
    return f(chute), chute       
```

```python
hooke_jeeves_1d(f, 0.1, 0.1, 0.0001, 0.00001, m.inf)
```

```python
a = 1.26105
b = 0.5340
c = 0.491092
d = 2.383
e = 12.1
f = 2.5209
g = -7.6
h = -9.89

beta = lambda pH, ext: a + b*ext + c*pH + d*ext*pH + e*ext**2 + f*pH**2 + g*ext**2*pH + h*ext*pH**2

import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from mpl_toolkits.mplot3d.axes3d import Axes3D
import numpy as np

pH = np.linspace(0.5, 1.5, 100)
ext = np.linspace(0.17, 0.35, 100)
pH_grid, ext_grid = np.meshgrid(pH, ext)
beta_grid = beta(pH_grid, ext_grid)

fig = plt.figure(figsize=(20,6))    
ax1 = fig.add_subplot(1, 2, 1, projection='3d')
p1 = ax1.plot_surface(pH_grid, ext_grid*100, beta_grid, cmap=mpl.cm.twilight, rstride=4, cstride=4, linewidth=0)
cset = ax1.contour(pH_grid, ext_grid*100, beta_grid, offset=1.2, zdir='z', cmap=mpl.cm.twilight_shifted)
cset = ax1.contour(pH_grid, ext_grid*100, beta_grid, offset=0.4, zdir='x', cmap=mpl.cm.twilight_shifted)
cset = ax1.contour(pH_grid, ext_grid*100, beta_grid, offset=40, zdir='y', cmap=mpl.cm.twilight_shifted)
ax1.set_xlabel('pH')
ax1.set_ylabel('% Extratante')
ax1.set_zlabel('Beta')
cb1 = fig.colorbar(p1, shrink=0.5)
```

```python
def acha_maior(f1, f2, f3, f4, f5):
    if f1 >= f2 and f1 >= f3 and f1 >= f4 and f1 >= f5:
        return f1
    elif f2 >= f3 and f2 >= f4 and f2 >= f5:
        return f2
    elif f3 >= f4 and f3 >= f5:
        return f3
    elif f4 >= f5:
        return f4
    else:
        return f5
```

```python
def hooke_jeeves_2d(z, chute_x, incremento_x, tolerancia_x, lim_inf_x, lim_sup_x,
                    chute_y, incremento_y, tolerancia_y, lim_inf_y, lim_sup_y):
        
    while (incremento_x > tolerancia_x or incremento_y > tolerancia_y):
        
        f1 = z(chute_x, chute_y)
        
        if chute_x + incremento_x <= lim_sup_x:
            f2 = z(chute_x + incremento_x, chute_y)
        else:
            f2 = -m.inf
            
        if chute_x - incremento_x >= lim_inf_x:
            f3 = z(chute_x - incremento_x, chute_y)
        else:
            f3 = -m.inf
            
        if chute_y + incremento_y <= lim_sup_y:            
            f4 = z(chute_x, chute_y + incremento_y)
        else:
            f4 = -m.inf
            
        if chute_y - incremento_y >= lim_inf_y:
            f5 = z(chute_x, chute_y - incremento_y)
        else:
            f5 = -m.inf
        
        maior = acha_maior (f1, f2, f3, f4, f5)
        #note p/ fazer um Hooke-Jeeves que acha o mínimo, eu só deveria trocar essa função para uma "acha_menor".
        #e trocar "-m.inf" por "m.inf"
        
        if maior == f1:            
            incremento_x /= 2
            incremento_y /= 2
            
        else:
            if maior == f2:
                chute_x += incremento_x
            elif maior == f3:
                chute_x -= incremento_x
            elif maior == f4:
                chute_y += incremento_y
            else:
                chute_y -= incremento_y
    
    return f1, chute_x, chute_y  
```
