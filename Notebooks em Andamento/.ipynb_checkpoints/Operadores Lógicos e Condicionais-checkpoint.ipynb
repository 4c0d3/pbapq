## Problema 1:

Em modelagem, a estabilidade de um sistema pode ser prevista em termos dos auto-valores da matriz que multiplica o vetor de variáveis de estado do sistema. Caso a parte real de ambos os auto-valores encontrados seja negativa, o sistema é estável; do contrário, é instável. Um sistema estável é aquele que alcança o equilíbrio, um sistema instável não o faz.

A definição de um autovalor é tal que:

$$ Ax = \lambda Ix$$

Onde A é a matriz de interesse. I é a matriz identidade e $\lambda$ é o autovalor, para os casos em que desconsiderando a solução trivial em que x é um vetor de de zeros.

Os autovalores podem ser obtidos da seguinte forma:

$$ |A - \lambda I| = 0$$

Lembrando que "| |" é a notação para determinante.

Seja A uma matriz 2 x 2, da forma:

$$ A = \begin{bmatrix}
        x_{11} & x_{12}\\
        x_{21} & x_{22}\\
       \end{bmatrix}$$

Então:

$$ \lambda I = \begin{bmatrix}
                \lambda & 0\\
                0 & \lambda\\
               \end{bmatrix}$$

E, portanto:

$$ |A - \lambda I| = \begin{vmatrix}
                  x_{11} - \lambda & x_{12}\\
                  x_{21} & x_{22} - \lambda\\
                  \end{vmatrix} = 0 $$
                  
O determinante é usado, então, para determinar $\lambda$:

$$ \lambda^2 - (x_{11} + x_{22}) \lambda + x_{11} x_{22} - x_{12} x_{21} = 0$$  

Um dado sistema tem:

$$ A = \begin{bmatrix}
        - 4 & 3\\
        - 5 & - 6\\
       \end{bmatrix}$$
       
Crie um algoritmo que calcule os autovalores de uma matriz 2 x 2 qualquer e, a partir disso, seja capaz de concluir sobre a estabilidade de um sistema. Use esse algoritmo para dizer se o sistema descrito por A é estável.

#### Passo 2: Crie uma condicional que receba os autovalores, e que retorne estável caso a parte real de ambos seja negativa e instável caso não sejam.

```python
x11 = -4
x12 = 3
x21 = -5
x22 = -6

a = 1
b = -(x11 + x22)
c = x11*x22 - x12*x21

def bhaskara(a, b, c):
    
    delta = b**2 - 4*a*c          
    x1 = (-b + delta**0.5)/2/a
    x2 = (-b - delta**0.5)/2/a
        
    if delta >= 0:
        return x1, x2
    # para o fim específico dessa função, nós não queremos que ela dê um retorno só, p/ qdo delta for 0.
    # outras ferramentas p/ resolver este problema serão vistas no curso. 
        """if delta > 0:
            return x1, x2
        else:
            return x1"""
    else:
        return x1.real, x2.real    
        
def isStable():
    lambda1, lambda2 = bhaskara(a, b, c)
    if lambda1 <=0 and lambda2 <= 0:
        return True
    else:
        return False

isStable()
```

## Problema 3:

Crie uma função que converte qualquer temperatura numa dada escala para a outra escala:

|Escala Inicial|Escala Final| Fórmula|
|:--------:|:--------:|:----------------------------:|
|Fahrenheit|Celsius   | ºC = (ºF - 32) / 1.8         |
|Fahrenheit|Kelvin    |	K = (ºF + 459.67) / 1.8      |
|Fahrenheit|Rankine   | Ra =  ºF + 459.67            |
|Celsius   |Fahrenheit| ºF =  ºC × 1.8 + 32          |
|Celsius   |Kelvin    |	K = ºC + 273.15              |
|Celsius   |Rankine   | Ra =  ºC × 1.8 + 32 + 459.67 |
|Kelvin    |Celsius   | ºC = K - 273.15              |
|Kelvin    |Fahrenheit| ºF = K × 1.8 - 459.67        |
|Kelvin    |Rankine   | Ra = K × 1.8                 |
|Rankine   |Celsius   | ºC = (Ra - 32 - 459.67) / 1.8|
|Rankine   |Fahrenheit| ºF =  Ra - 459.67            |
|Rankine   |Kelvin    | K =  Ra / 1.8                |


```python
def conversor_temperatura(t, escala_i, escala_f):
    
    if escala_i == escala_f:
        return t
    
    else:
        if escala_i == 'celsius':
            if escala_f == 'kelvin':
                return t + 273
            elif escala_f == 'fahrenheit':
                return t * 9/5 + 32
            else:
                return t * 9/5 + 491.67
            
        elif escala_i == 'kelvin':
            if escala_f == 'celsius':
                return t - 273.15
            elif escala_f == 'fahrenheit':
                return (t - 273.15) * 9/5 + 32
            else:
                return t * 9/5
            
        elif escala_i == 'fahrenheit':
            if escala_f == 'celsius':
                return (t - 32) * 5/9
            elif escala_f == 'kelvin':
                return (t - 32) * 5/9 + 273.15
            else:
                return t + 459.67
        
        else:
            if escala_f == 'celsius':
                return t * 5/9 - 273.15
            elif escala_f == 'kelvin':
                return t * 5/9
            else:
                return t - 459.67  
```

A amônia nas CNTP não é inflamável, mas é extremamente volátil e tóxica. A massa de amônia vazada, obtida anteriormente, nos será muito importante para esse cálculo da dispersão do gás, que irá se espalhar na direção do vento, formando uma pluma, como na imagem abaixo (lá na frente, faremos a nossa própria pluma). 



No nosso caso, a pluma é formada a partir do chão ($altura = 0$) e estamos interessados nos potenciais danos aos operadores no local do vazamento e população local na direção da pluma. Logo, usaremos uma hipótese simplificadora para ignorarmos a altura nas equações da pluma.

A primeira coisa que precisamos calcular é a taxa de evaporação da amônia na poça, que é dada por:

$$ Q = { MKAP^{sat} \over RT } $$

Onde:
- Q é a taxa de evaporação em massa/tempo.
- M é a massa molar da amônia.
- A é a área da poça = 1000 m².
- K é a constante de transferência de massa da amônia.
- $P^{sat}$ é a pressão de saturação da amônia.
- R é a constante dos gases ideais.
- T é a temperatura (em escala absoluta).

Algumas observações importantes:

- A área da poça é constante ao longo de todo vazamento (isso é o mesmo que dizer que desde $t = 0$ até o momento que o vazamento cessa, a amônia vaza num local cuja área é a mesma para qualquer profundidade da poça. Tanques costumam ficar cercados por diques (foto abaixo), o que faria do local de vazamento um paralelepípedo, por exemplo, atendendo o critério. Reatores não costumam estar entre diques, mas assumamos que a área de produção esteja; também aproximemos que a área é uniformemente preenchida desde $t = 0$.



- Dada a simplificação acima, Q foi considerado constante ao longo de todo o tempo. Note que a massa de amônia vazada não afeta em nada esse valor de Q, por conta dessa aproximação. Na prática, a quantidade de amônia vazada tem relação com a área da poça para volumes pequenos (insuficientes para gerar uma área do tamanho da área do dique) e grandes (que transbordariam o dique). Além disso, a massa afeta o tempo de evaporação para uma dada taxa de evaporação, claro.

K pode ser obtido pela relação:

$$ K = K_0 \cdot \bigg ({M_0 \over M} \bigg)^{1/3}$$ 

Onde $M_0$ e $K_0$ são as mesmas propriedades M (massa molar) e K (constante de transferência de massa) de um composto de referência, geralmente a água.

$K_{H_2O} = 0,83 cm/s$

$P^{sat}$, em Pascal, pode ser obtido pela relação (Perry $8^{th}$ Ed.):

$$ \Large {P^{sat} = e^{(C_1 + C_2/T + C_3 ln(T) + C_4 T^{C_5})}} $$

Para $NH_3$, temos:

$C_1 = 90$  
$C_2 = -4669,7$  
$C_3 = -11,607$   
$C_4 = 1,72 \cdot 10^{-2}$  
$C_5 = 1$

Nós podemos **estimar** a concentração de amônia numa dada posição na direção do vento, dadas as observações já feitas utilizando o modelo Pasquill-Gifford. Note que embora a área tenha sido importante para determinar Q; na hora de projetar a pluma, a determinação da concentração de amônia no ar assumirá para a localização cartesiana do ponto que a poça é um ponto na origem.

$$ \large{ C_{x, y, (z=0)} = {Q \over \pi \sigma_y \sigma_z u} \cdot
 e^{\bigg[{-1 \over 2} \bigg ({y^2 \over \sigma_y^2}\bigg)\bigg]} }$$

- x e y são as coordenadas cartesianas.

```python
#------------Dados Metereológicos da Fafen-SE.------------------
T = 24 #ºC
T = conversor_temperatura(T, 'celsius', 'kelvin') #A variável T está sendo sobreescrita, usando a T velha como argumento.
u = 0.8 #m/s ---> velocidade do vento
isRural = False
horario = 7
condTempo = 'completamente nublado'

#------------Dados para os modelos------------------

A = 1000 #m²
R = 8.314 #J/K/mol

K0 = 0.83 #cm/s
M0 = 18 #g/mol
M = 17 #g/mol

c1= 90 
c2= -4669.7 
c3= -11.607 
c4= 1.72e-2 
c5= 1 

#----------------Conversões SI----------------------
K0 = K0/100 #m/s
M0 = M0/1000 #kg/mol
M = M/1000 #kg/mol
```

```python
# uma regra prática: se você não pretende mexer num certo valor (como numa constante),
# deixe ele apenas no escopo global (definido fora de uma função) e não coloque como argumento da função.
# quando formos plotar a nossa pluma, o único valor que irá ficar mudando continuamente será "x" e "y".
K = lambda: K0 * (M0/M)**(1/3)
Psat = lambda: e**(c1 + c2/T + c3*ln(T) + c4*T**c5)
Q = lambda: M*K()*A*Psat()/R/T
```

```python
def classeEstabilidade():
    """ 
    assegurar que se caso dia ou noite seja escolhido, informação correta sobre o tempo é inputada.
    evitando algo como "noite com insolação forte" ou 
    "dia parcialmente nublado (que embora possível, deve ser colocado em termos de insolação)". 
    """
    assert horario >= 0 and horario <= 24
    assert (
        horario >= 7 and horario <= 17 and 
        (condTempo == 'forte' or condTempo == 'medio' or condTempo == 'fraco' or condTempo == 'completamente nublado') or 
        
        (horario >= 0 and horario <= 6 or horario >= 18 and horario <= 24) and
        (condTempo == 'parcialmente nublado' or condTempo == 'claro' or condTempo == 'completamente nublado')
    )
    #-------------------------------------------------------------------------------------------------------------
    
    if (horario >= 6 and horario <= 7 or horario >= 17 and horario <= 18) or condTempo == 'completamente nublado':
        return 'D'
    
    elif horario >= 7 and horario <= 17:
        if condTempo == 'forte':
            if u <= 3:
                return 'A'
            elif u <= 4:
                return 'B'
            else:
                return 'C'                
            
        elif condTempo == 'medio':
            if u <= 2:
                return 'A'
            elif u <= 3:
                return 'B'
            elif u <= 6:
                return 'C'
            else:
                return 'D'
            
        else:
            if u <= 2:
                return 'B'
            elif u <= 4:
                return 'C'
            else:
                return 'D'
            
    else:
        if condTempo == 'parcialmente nublado':
            if u <= 2:
                return 'F'
            elif u <= 3:
                return 'E'
            else:
                return 'D'
        else:
            if u <= 3:
                return 'F'
            elif u <= 4:
                return 'E'
            else:
                return 'D' 
```

```python
def sigma(x):
    assert type(isRural) == bool # isRural deve ser um input de True ou False.
    classe_estabilidade = classeEstabilidade().capitalize() 
    """
    um truquezinho p/ garantir que tanto letras maiúsculas e minúsculas serão reconhecidas.
    Ele passa tudo para maiúsculo.
    """   
    if isRural:
        if classe_estabilidade == 'A':
            sigmay = 0.22*x*(1 + 0.0001*x)**(-1/2)
            sigmaz = 0.2*x
        elif classe_estabilidade == 'B':
            sigmay = 0.16*x*(1 + 0.0001*x)**(-1/2)
            sigmaz = 0.12*x
        elif classe_estabilidade == 'C':
            sigmay = 0.11*x*(1 + 0.0001*x)**(-1/2)
            sigmaz = 0.08*x*(1 + 0.0002*x)**(-1/2)
        elif classe_estabilidade == 'D':
            sigmay = 0.08*x*(1 + 0.0001*x)**(-1/2)
            sigmaz = 0.06*x*(1 + 0.00015*x)**(-1/2)
        elif classe_estabilidade == 'E':
            sigmay = 0.06*x*(1 + 0.0001*x)**(-1/2)
            sigmaz = 0.03*x*(1 + 0.0003*x)**-1
        else:
            sigmay = 0.04*x*(1 + 0.0001*x)**(-1/2)
            sigmaz = 0.016*x*(1 + 0.0003*x)**-1
            
    else:
        if classe_estabilidade == 'A' or classe_estabilidade == 'B':
            sigmay = 0.32*x*(1 + 0.0004*x)**(-1/2)
            sigmaz = 0.24*x*(1 + 0.0001*x)**(1/2)
        elif classe_estabilidade == 'C':
            sigmay = 0.22*x*(1 + 0.0004*x)**(-1/2)
            sigmaz = 0.2*x
        elif classe_estabilidade == 'D':
            sigmay = 0.16*x*(1 + 0.0004*x)**(-1/2)
            sigmaz = 0.14*x*(1 + 0.0003*x)**(-1/2)
        else:
            sigmay = 0.11*x*(1 + 0.0004*x)**(-1/2)
            sigmaz = 0.08*x*(1 + 0.0015*x)**(-1/2)
            
    return sigmay, sigmaz    
```

```python
def concentracao(x, y):
    """
    experimente o modelo para valores muito pequenos de x e veja que ele perde completamente o significado físico.
    """
    k = K()    
    psat = Psat()
    q = Q()
    classe_estabilidade = classeEstabilidade()
    sigmay, sigmaz = sigma(x)
    
    conc = q/pi/u/sigmay/sigmaz * e**(-1/2 * y**2/sigmay**2) *1e6
    return conc  
```

```python
import numpy as np
from numpy import ma
import matplotlib.pyplot as plt
from matplotlib import ticker, cm
import matplotlib.colors as colors
import matplotlib.cbook as cbook

aegl1 = 30 #mg/m³
aegl2 = 160 #mg/m³
aegl3 = 1100 #mg/m³

x1s = [i for i in np.logspace(0, 4, 300)]
y1s = [i for i in np.logspace(0, 3, 300)]
y2s = [-i for i in np.logspace(0, 3, 300)]

X1s, Y1s = np.meshgrid(x1s, y1s)
X2s, Y2s = np.meshgrid(x1s, y2s)
Concentracoes1 = concentracao(X1s, Y1s)
Concentracoes2 = concentracao(X2s, Y2s)
Concentracoes1 = ma.masked_where(Concentracoes1 < 10, Concentracoes1)
Concentracoes2 = ma.masked_where(Concentracoes2 < 10, Concentracoes2)
gambiarra1, gambiarra2 = np.meshgrid(np.array([0, 1]), np.array([0, 1]))


fundo = plt.figure(figsize = (14, 10))
contorno = fundo.add_subplot(1, 1, 1)

gambiarra = contorno.contourf(gambiarra1, gambiarra2, gambiarra2, cmap = cm.Blues, 
                              levels = [aegl1, aegl2, aegl3])
metadeSuperior = contorno.contourf(X1s, Y1s, Concentracoes1, cmap = cm.Blues,
                                   norm=colors.LogNorm(vmin=aegl1, vmax=aegl3), levels = [aegl1, aegl2, aegl3, 1e6])
metadeInferior = contorno.contourf(X2s, Y2s, Concentracoes2, cmap = cm.Blues, 
                                   norm=colors.LogNorm(vmin=aegl1, vmax=aegl3), levels = [aegl1, aegl2, aegl3, 1e6])
contorno.set_ylabel('Y')
contorno.set_xlabel('X')
contorno.set_title("Azul Claro > AEGL1 \t \t \t Azul > AEGL2  \t \t \t Azul Escuro > AEGL3")
contorno.set_ylim(-2000, 2000)
contorno.set_xlim(0, 10000)
cb = fundo.colorbar(gambiarra, ax=contorno)
```
